```
import face_recognition as fr
```

```
image = fr.load_image_file('/content/drive/My Drive/Study/fam.jpg')
```

```
from google.colab.patches import cv2_imshow
```

```
cv2_imshow(image)
```

```
locs = fr.face_locations(image) #facelocation
```

```
print(len(locs), "faces detected") #coordinates of each face as a square
```

```
print(locs)
```

```
import cv2 #computer vison (CV)
```

```
for face in locs:
  cv2.rectangle(image, (face[3], face[0]), (face[1], face[2]), (9,255,0),2) #coordinate for 4 points left, top, bottom, right
```

```
cv2_imshow(image)
```
